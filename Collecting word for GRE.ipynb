```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
path = "C:\Driver\chromedriver.exe"
driver = webdriver.Chrome(executable_path=path)
driver.get("https://www.graduateshotline.com/gre-word-list.html#x2")
driver.maximize_window()
```

```python
word = []
define =[]
words = driver.find_elements(By.TAG_NAME,'tr')
# for word in words:
#     print(word.text)
# #     # prin_word = driver.find_element(By.XPATH,"./td[1]").text
# #     # print(prin_word)
for word_ in words:
    w = word_.find_element(By.XPATH,"./td[1]").text
    print(w)
    word.append(w)

    deff= word_.find_element(By.XPATH,"./td[2]").text
    print(deff)
    define.append(deff)
```

```python
gre_book = pd.DataFrame({
    'Word': word,
    'Definition':define
})
```

```python
gre_book.to_excel("GRE_Wordsmurtuza.xlsx",index=False)
```
